Now we can use awkward arrays to manipulate the information we have grabbed from the ROOT file. Awkward arrays are jagged arrays which allow us to operate on all of the events at one time rather than looping through all of the events. For this example, we will grabt the muons and the electrons and make some simple distributions to show the basic awkward tools

```python
import awkward as ak
import vector
vector.register_awkward()
#table = awkward.Table(branches)
```

```python
Muons = ak.zip({
      "pt": arrays['Muon_pt'],
      "eta": arrays['Muon_eta'],
      "phi": arrays['Muon_phi'],
      "mass": arrays['Muon_mass']
}, with_name="Momentum4D") 
```

We can now index this collection to see what is happening. For example we can see the pts of the muons in the first event (Muons.pt[0]) or in the last event (Muons.pt[-1]). Lets also using awkard to count the number of muons in each event and see if it matches with the nMuon variable stored in the NanoAOD file.

```python
Muons.pt[0]
```

```python
Muons.pt[-1]
```

```python
ak.num(Muons)
```

```python
ak.num(Muons) - arrays['nMuon']
```

We can also add basic selections to these collections as well. Lets take the muons as an example for now and add some basic selections.

```python
cut = (arrays['Muon_pt'] > 10) & \
            (abs(arrays['Muon_eta']) <= 2.4) & \
            (arrays['Muon_mediumId'] == 1) 
Muons = Muons[cut]
```

Now that we have a collection of Muons that we are interested in we can make some simple histograms using Boost Histogram. For this, we will also use mplhep with CMS style to make our plots. 

```python
import mplhep as hep
import hist

#hep.cms.label(data=False, year='2018')
hep.style.use("CMS")
# {"ALICE" | "ATLAS" | "CMS" | "LHCb1" | "LHCb2"}
hep.cms.label(data=False, lumi=60, year=2018)

h = hist.Hist(hist.axes.Regular(10,0,10,name="nMuons"))
h.fill(nMuons=ak.num(Muons))
h.plot()
```

This is just the muon collection. Let's make a collection that combines the electrons and muons together. Then we can look at the total number of leptons (excluding tau) in an event.

```python
Electrons = ak.zip({
      "pt": arrays['Electron_pt'],
      "eta": arrays['Electron_eta'],
      "phi": arrays['Electron_phi'],
      "mass": arrays['Electron_mass']
}, with_name="Momentum4D") 
cut_ele = (arrays['Electron_pt'] > 10) & \
            (abs(arrays['Electron_eta']) <= 2.4) & \
            (arrays['Electron_mvaFall17V2Iso_WP90'] == 1) 
Electrons = Electrons[cut_ele]
```

```python
nLepton = ak.num(Muons) + ak.num(Electrons)

hep.cms.label(data=False, lumi=60, year=2018)
hlep = hist.Hist(hist.axes.Regular(10,0,10,name="nLeptons"))
hlep.fill(nLeptons=ak.num(Muons) + ak.num(Electrons))
hlep.plot()
```

That was done by just adding the numbers together. Lets actually form a new collection that is the full combination of Muons and Electrons. Then we can comapre the number of leptons we see from that collection. We can also look at the events with at least 2 leptons and look at variables associated with the first 2 leptons.

```python
Leptons = ak.concatenate([Muons, Electrons], axis=1)

hep.cms.label(data=False, lumi=60, year=2018)
hlep2 = hist.Hist(hist.axes.Regular(10,0,10,name="nLeptons"))
hlep2.fill(nLeptons=ak.num(Leptons))
hlep2.plot()
#If you want to compare to the old plot
#hlep.plot()
```

```python
Leptons = Leptons[ak.num(Leptons)>1]
Leading_lepton = Leptons[:,0]
Trailing_lepton = Leptons[:,1]
```

```python
hep.cms.label(data=False, lumi=60, year=2018)
hpt = hist.Hist(hist.axes.Regular(25,0,100,name="Lepton_pt"))
hpt.fill(Lepton_pt=Leading_lepton.pt)
hpt.plot()
hpt2 = hist.Hist(hist.axes.Regular(25,0,100,name="Lepton_pt"))
hpt2.fill(Lepton_pt=Trailing_lepton.pt)
hpt2.plot()
```
